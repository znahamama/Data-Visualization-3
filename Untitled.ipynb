```python
import matplotlib.pyplot as plt

import pandas as pd

from matplotlib.gridspec import GridSpec

from pywaffle import Waffle

from matplotlib.patches import Patch
```

```python
df = pd.read_csv('consoles.csv')
df1 =df.groupby('Company').sum()

df2= df.groupby('Platform').sum().reset_index()
df2=df2.sort_values(by=['Company','Units sold'],ascending=True)

nintendo_least= df2[df2["Company"]=='Nintendo'].head(4)

df2['Label'] = df2['Platform']
df2.loc[df2['Platform'].isin(nintendo_least['Platform']), 'Label'] = ''  

colors = {'Nintendo': '#EA6727', 'Sony': '#83C143', 'Microsoft': '#5599CC'}
df2['Color'] = df2['Company'].map(colors)
inside_color= ['#156EAF', '#DB2018', '#56B45B']

fig, ax = plt.subplots(figsize=(9, 6))

outer, _ = ax.pie(df2['Units sold'],labels=df2['Label'],colors=df2['Color'],radius=1.3,wedgeprops={'edgecolor': 'white'})
inner, _ = ax.pie(df1['Units sold'],labels=df1.index,colors=inside_color,labeldistance= 0.4,textprops={'color': 'white'},radius=0.9,wedgeprops={'edgecolor': 'white'})

ax.set_title('Best Selling Video Game Consoles',pad=50)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
from matplotlib.gridspec import GridSpec
from matplotlib.lines import Line2D  

df = pd.read_csv('gen-ai.csv')

df = df.drop("total", axis=1).T
df.columns = df.iloc[0]
df = df.drop("occupation")
job_list = df.columns.values.tolist()
df = pd.DataFrame(df, columns=job_list)

colors = ['#065B41', '#36D4AC', '#D3D3D3']

legend_labels = ['Man', 'Woman', 'Ambiguous']
legend_elements = [
    Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markersize=15, label=label)
    for color, label in zip(colors, legend_labels)
]

num_jobs = len(job_list)
fig = plt.figure(figsize=(16, 8))
gs = GridSpec(5, (num_jobs + 1) // 2, height_ratios=[0.25, 0.1, 1, 0.025, 1])

legend_ax = fig.add_subplot(gs[0, :])
legend_ax.axis('off')  

legend = legend_ax.legend(handles=legend_elements,loc='best',bbox_to_anchor=(0.4, 0.5),ncol=len(legend_elements),frameon=False,fontsize=14,handletextpad=0.1)

fig.text(0.68, 0.87, "All but two images for \nthe keyword ''Engineer'' \nwere of perceived men", fontsize=14, ha='left')
fig.text(0.01, 0.903, "Perceived Gender:", fontsize=14, ha='left')
fig.text(0.01, 0.84, "High-paying occupations", fontsize=14, fontweight="bold", ha='left')

#waffle graphs for high paying occupations
axes_high = [fig.add_subplot(gs[2, i]) for i in range((num_jobs + 1) // 2)]
for idx, job in enumerate(job_list[:(num_jobs + 1) // 2]):
    ax = axes_high[idx]
    Waffle.make_waffle(rows=20,columns=15,values=df.iloc[:, idx],labels=None,vertical=True,title={'label': job.upper(), 'loc': 'left', 'fontsize': 12},colors=colors,ax=ax,)
    ax.axis("off") 

fig.text(0.01, 0.4, "Low-paying occupations", fontsize=14, fontweight="bold", ha='left')

#waffle graphs for low paying occupations
axes_low = [fig.add_subplot(gs[4, i]) for i in range((num_jobs + 1) // 2)]
start_idx = (num_jobs + 1) // 2  
for idx, job in enumerate(job_list[start_idx:]):
    ax = axes_low[idx]
    Waffle.make_waffle(rows=20,columns=15,values=df.iloc[:, idx + start_idx],labels=None,vertical=True,title={'label': job.upper(), 'loc': 'left', 'fontsize': 12},colors=colors,ax=ax,)
    ax.axis("off") 
    
plt.tight_layout()

#seperate axis to draw the arrow
ax_arrow = fig.add_axes((0.83,0.8,0.153,0.15))
for side in ['left', 'bottom', 'right', 'top']: 
    ax_arrow.spines [side].set_visible(False)
    
ax_arrow.set_xticks( [])
ax_arrow.set_yticks([])

#coordinates to form arrow
ax_arrow.plot([0, 0.15, 0.35, 0.55, 0.7, 0.82, 0.9, 0.96, 1, 1], [0.8, 0.95, 1, 0.95, 0.85, 0.72, 0.55, 0.4, 0.2, 0], color="black")
ax_arrow.scatter(0.996, 0, color="black", linewidth=0.5, marker='v', s=50)

plt.show()
```
